```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt

headbrain = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/headbrain.csv")
print(headbrain.head())
```

```python
headbrain
```

```python
#Linear reln bw headsize and brainweight
X=headbrain['Head Size(cm^3)'].values
Y = headbrain['Brain Weight(grams)'].values
```

```python
np.corrcoef(X, Y)
```

```python
plt.scatter(X, Y, c='blue', label='Data points')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate m and c
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

# Printing coefficients
print("coefficients for regression",b1, b0)
```

```python
# Ploting Line
plt.plot(X, Y, color='red', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='green', label='Scatter data')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
# Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = b0 + b1 * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score",r2)
```
